```python
import pandas as pd

# Load the zipped CSV file directly into a pandas DataFrame
df = pd.read_csv('/content/Twitter_Data.csv.zip')

# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
# Display general information about the DataFrame, including data types and non-null values
df.info()
```

```python
# Check for missing values
print('Missing values before dropping:')
display(df.isnull().sum())

# Drop rows with any missing values as they are few and might cause issues in sentiment analysis.
df.dropna(inplace=True)

print('\nMissing values after dropping:')
display(df.isnull().sum())
```

```python
# Convert 'category' column to integer type as it represents discrete sentiment labels
df['category'] = df['category'].astype(int)

# Display the distribution of sentiment categories
print('\nDistribution of Sentiment Categories:')
display(df['category'].value_counts())
```
